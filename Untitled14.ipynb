```python
#원핫 인코딩 및 ndarray로의 변환
print("one-hot & ndarray")

from tensorflow.keras.utils import img_to_array, array_to_img#이미지를 배열로, 배열을 이미지로 변환하는 함수 임포트

train_x=np.asarray([img_to_array(array_to_img(img, scale=False).resize((224,224))) for img in train_x]).astype('uint8')#resize(224, 224)를 위해 배열로 잠시 바꾼뒤 색상정보 타입을 모델과의 호환을 위해 uint8으로 바꾼다.
test_x=np.asarray([img_to_array(array_to_img(img, scale=False).resize((224,224))) for img in test_x]).astype('uint8')

label_dict={"ur":0, "_u":1, "ul":2, "_r":3, "_p":4, "_l":5, "dr":6, "_d":7, "dl":8}#정수 인코딩 딕셔너리
label_dict_r={0:"ur", 1:"_u", 2:"ul", 3:"_r", 4:"_p", 5:"_l", 6:"dr", 7:"_d", 8:"dl"}
train_y=[label_dict[label] for label in train_y]
test_y=[label_dict[label] for label in test_y]
train_y=np.asarray(train_y)
test_y=np.asarray(test_y)

train_y=to_categorical(train_y)
test_y=to_categorical(test_y)
```

```python
import torch
from PIL import Image

#Dataset 클래스
class MyDataset(torch.utils.data.Dataset):
  def __init__(self, img_list, label_list, phase, transform):#(훈련용 이미지와 정답 리스트, train과 validation모드의 지정, 추가번형)
    self.img_list=img_list
    self.label_list=label_list
    self.phase=phase#transform람다에 전달할 인자.
    self.transform=transform

  def __len__(self):
    return len(self.img_list)#이미지 개수를 반환

  def __getitem__(self, index):
    img=self.img_list[index]#ndarray
    img=array_to_img(img, scale=False)

    transformed_img=self.transform(img, self.phase)#변환을 거친 뒤
    label=self.label_list[index]#해당 이미지 인덱스에 해당하는 정답을 가져옴

    return transformed_img, label#변환된 이미지와 정답을 리턴
```

```python
from torchvision import models, transforms

class MyTransform():
  def __init__(self):
    self.data_transform={
        'train': transforms.Compose([
            transforms.ToTensor()
        ]),

        'val': transforms.Compose([
            transforms.ToTensor()
        ])
    }

  def __call__(self, img, phase='train'):
    return self.data_transform[phase](img)
```

```python
#데이터셋 인스턴스 생성_입출력의 구분과 변형(전처리)를 위함
print("making instance of dataset...")

train_dataset=MyDataset(img_list=train_x, label_list=train_y, phase='train', transform=MyTransform())
val_dataset=MyDataset(img_list=test_x, label_list=test_y, phase="val", transform=MyTransform())
image_datasets={'train': train_dataset, 'val':val_dataset}
dataset_sizes={x: len(image_datasets[x]) for x in ['train', 'val']}
```

```python
#데이터로더 인스턴스 생성_배치크기와 셔플기능 추가를 위함
print('making instance of dataloader...')

batch_size=64
train_dataloader=torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_dataloader=torch.utils.data.DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
dataloaders_dict={"train": train_dataloader, "val": val_dataloader}
```

```python
import timm
import torch.nn as nn
#VIT모델
num_classes=128#nine gaze
vit_model=timm.create_model("vit_base_patch16_224", pretrained=False, num_classes=num_classes)

model = nn.Sequential(
    vit_model,
    nn.Linear(128, 32),
    nn.ReLU(),
    nn.Linear(32,9),
    nn.ReLU(),
    nn.Softmax()
)
```

```python
#손실함수, 역전파
import torch.optim as optim
import torch.nn as nn
from torch.optim import lr_scheduler

criterion=nn.CrossEntropyLoss()
optimizer=optim.AdamW(model.parameters())
#exp_lr_scheduler=lr_scheduler.StepLR(optimizer, step_size=7, gamma=0.1)
```

```python
#훈련시작
print("start training...")

device=torch.device('cuda:0' if torch.cuda.is_available() else "cpu")

epochs=50
best_acc=0
for epoch in range(epochs):
  print("{}/{} epoch running now".format(epoch+1, epochs))

  for phase in ['train', 'val']:
    if phase=='train':
      model.train()
    else:
      model.eval()

    running_loss=0.0
    running_correct=0

    for inputs, labels in dataloaders_dict[phase]:
      inputs=inputs.to(device)
      labels=labels.to(device)

      optimizer.zero_grad()

      with torch.set_grad_enabled(phase=='train'):
        outputs=model(inputs)
        preds=torch.argmax(outputs, dim=1)#최대값의 인덱스 값
        preds=to_categorical(preds, num_classes=9)#해당 값을 one-hot encoding

        loss=criterion(outputs, labels)

        if phase=='train':
          loss.backward()
          optimizer.step()

      running_loss+=loss.item()*inputs.size(0)
      running_corrects=0.0
      for i in range(len(preds)):
        if np.argmax(preds[i])==np.argmax(labels.data[i]):
          running_corrects+=1
      

    #if phase=='train':
      #exp_lr_scheduler.step()
    
    epoch_loss=running_loss/dataset_sizes[phase]
    epoch_acc=running_corrects/dataset_sizes[phase]
    print("{} Loss: {:.4f} Acc: {:.4f}".format(phase, epoch_loss, epoch_acc))

    #모델 저장
    if phase=='val' and epoch_acc>best_acc:
      best_acc=epoch_acc
      torch.save(model.state_dict(), 'best_model.pth')
```

```python
#시각화

def visualize_model(model, num_images=6):
  was_training=model.training
  model.eval()
  images_so_far=0
  fig=plt.figure()

  with torch.no_grad():
    for i, (inputs, labels) in enumerate(dataloaders_dict['val']):
      inputs=inputs.to(device)
      labels=labels.to(device)

      outputs=model(inputs)
      _, preds=torch.max(outputs, 1)

      for j in range(inputs.size()[0]):
        images_so_far+=1
        ax=plt.subplot(num_images//2, 2, images_so_far)
        ax.axis('off')
        ax.set_title('predicted: {}'.format(label_dict[preds[j]]))
        plt.imshow(inputs.cpu().data[j])

        if images_so_far==num_images:
          model.train(mode=was_training)
          return

      model.train(mode=was_training)
```
